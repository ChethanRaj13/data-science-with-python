```python
import json

def load_data(filename):
    with open(filename,'r') as f:
        return json.load(f)

def give_people_suggestions(user_id, data):
    user_friends = {}
    for user in data["users"]:
        user_friends[user['id']] = set(user['friends'])

    if user_id not in user_friends:
        return []

    direct_friend = user_friends[user_id]
    suggested = {}
    for friend in direct_friend:
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friend:
                suggested[mutual] = suggested.get(mutual,0) + 1
                sorted_suggestions = sorted(suggested.items(), key=lambda x: x[1], reverse=True)
    return [user_id for user_id, _ in sorted_suggestions]

data = load_data("people.json")
user_id = 1
suggest = give_people_suggestions(user_id, data)

print(suggest)
```
