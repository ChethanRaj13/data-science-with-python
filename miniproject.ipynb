```python
import json;
```

```python
#take the file name and load its data and return it 
def load_data(filename):
    with open(filename,"r") as file:
        data = json.load(file)
    return data
```

```python
#display the data 
def display_data(data):
    print(f"Users and their friend with liked pages\n");
    for user in data["users"]:
        print(f"ID: {user['id']} - {user['name']} is friends with {user['friends']} and likes the page {user['liked_pages']}")   

    print(f"\nthe pages are:")
    for page in data["pages"]:
        print(f"Page ID: {page['id']} - {page['name']}")
```
